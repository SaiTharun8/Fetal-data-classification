# Pre-processing

```python
import missingno as msgno
```

```python
msgno.bar(fetal_data)
#plotting null values in bar whether to notice any variable has null values.
```

```python
from sklearn.preprocessing import MinMaxScaler
#importing minmax scaler for feature scaling.
```

```python
for x in fetal_data.columns:
    freq=fetal_data[x].value_counts()
    print(freq)
#getting all the value counts of each varible in the data, so as to find any categorical variable exist.
```

```python
mm=MinMaxScaler()
```

```python
fetal_mm=mm.fit_transform(fetal_data.iloc[:,0:21])
#scaling the columns except the target variable.
```

```python
fetal_data.columns
```

```python
fetal_mm=pd.DataFrame(fetal_mm,columns=['baseline value', 'accelerations', 'fetal_movement',
       'uterine_contractions', 'light_decelerations', 'severe_decelerations',
       'prolongued_decelerations', 'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'percentage_of_time_with_abnormal_long_term_variability',
       'mean_value_of_long_term_variability', 'histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency'])
```

```python
fetal_mm=fetal_mm.reset_index()
#getting the scaled feartures into proper index. 
```

```python
print( fetal_data.head())
print(  fetal_mm.head())
#before scaling the features and after scaling the features
```

```python
x=fetal_mm.drop('index',axis=1)
y=fetal_data.fetal_health
#bringing into x and y format.
```

```python
y.value_counts()
```

```python
y.value_counts().plot(kind='bar')
#this shows that the data is imbalanced, which becomes problem in predicting the variable.
```

```python
from imblearn.over_sampling import SMOTEN
#importing SMOTEN library for categorical data.
```

```python
smote=SMOTEN(random_state=41)
x_smote,y_smote=smote.fit_resample(x,y)
```

```python
y_smote.value_counts().plot(kind='bar')
#the vaues are now balanced and can be used for model building.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
logit=LogisticRegression(max_iter=3000)
```

```python
logit_model=logit.fit(x_smote,y_smote)
```

```python
logit_model.score(x_smote,y_smote)
```

```python
cross_val_score(logit_model,x_smote,y_smote)
```

```python
np.mean([0.81067472, 0.88922457, 0.93152064, 0.91137966, 0.81470292])
```

```python
logit_predict=logit.predict(x_smote)
```

```python
pd.crosstab(y_smote,logit_predict)
```

```python
ConfusionMatrixDisplay.from_predictions(y_smote,logit_predict)
```

```python
print(classification_report(y_smote,logit_predict))
#all the remaining metrics are also for logistic regression.
```
